#**Constrainte Satisfaction Problem**

**GT = Generate & Test**

```python
import itertools

regions = ['A', 'T', 'M', 'F', 'U', 'D']
couleurs = ['Rouge', 'Vert', 'Bleu']
contraintes = [
    ('A', 'T'), ('A', 'M'),
    ('A', 'F'), ('A', 'U'),
    ('T', 'F'), ('T', 'D'),
    ('T', 'M'), ('M','U'), ('M','D'), ('F','U')
]
```

```python
def est_valide(assignation):
    for r1, r2 in contraintes:
        if assignation[r1] == assignation[r2]:
            return False
    return True

def generate_and_test():
    nb_combinaisons = 0

    for combo in itertools.product(couleurs, repeat=len(regions)):
        nb_combinaisons += 1

        assignation = dict(zip(regions, combo))

        if est_valide(assignation):
            # print(f"Solution trouvée après {nb_combinaisons} tentatives !")
            return assignation

    print("Aucune solution trouvée.")
    return None

# Exécution
solution = generate_and_test()
if solution:
    for reg, col in solution.items():
        print(f"{reg} : {col}")
```

**BackTracking**

```python
def est_sur(region, couleur, assignation):
    for (r1, r2) in contraintes:
        if r1 == region and r2 in assignation:
            if assignation[r2] == couleur:
                return False
        if r2 == region and r1 in assignation:
            if assignation[r1] == couleur:
                return False
    return True

def backtracking(assignation={}):
    if len(assignation) == len(regions):
        return assignation

    region = [r for r in regions if r not in assignation][0]

    for couleur in couleurs:
        if est_sur(region, couleur, assignation):
            assignation[region] = couleur
            resultat = backtracking(assignation)
            if resultat is not None:
                return resultat

            del assignation[region]

    return None

solution = backtracking()

if solution:
    print("Backtracking :")
    for reg, col in solution.items():
        print(f"  {reg} : {col}")
else:
    print("Pas de solution possible.")
```

**Forward Checking**

```python
domaines_initiaux = {r: list(couleurs) for r in regions}

def obtenir_voisins(v):
    voisins = []
    for (r1, r2) in contraintes:
        if r1 == v: voisins.append(r2)
        elif r2 == v: voisins.append(r1)
    return voisins

def forward_checking(assignation, domaines):
    if len(assignation) == len(regions):
        return assignation

    v = [r for r in regions if r not in assignation][0]

    for couleur in list(domaines[v]):

        nouveaux_domaines = (domaines)
        assignation[v] = couleur

        echec_anticipe = False
        for voisin in obtenir_voisins(v):
            if voisin not in assignation:
                if couleur in nouveaux_domaines[voisin]:
                    nouveaux_domaines[voisin].remove(couleur)

                if not nouveaux_domaines[voisin]:
                    echec_anticipe = True
                    break

        if not echec_anticipe:
            resultat = forward_checking(assignation, nouveaux_domaines)
            if resultat is not None:
                return resultat

        del assignation[v]

    return None

solution = forward_checking({}, domaines_initiaux)

if solution:
    print("Forward Checking :")
    for reg in sorted(solution.keys()):
        print(f"  Province {reg} : {solution[reg]}")
else:
    print("Pas de solution.")
```
